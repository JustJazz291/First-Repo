```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('/content/British Airways Summer Schedule Dataset - Forage Data Science Task 1.csv')
```

```python
dataset.head()
```

```python
# Example logic (customize as per your rules)
def categorize_flight(row):
    time = int(row['FLIGHT_TIME'].split(':')[0])
    dest = row['ARRIVAL_REGION']

    if dest in ['Europe'] and time < 12: # Assuming 'Europe' contains Madrid, Paris, Berlin
        return 'Short-haul AM'
    elif dest in ['North America'] and time < 12: # Assuming 'North America' contains Larnaca, Cairo - this may need adjustment based on actual data
        return 'Mid-haul AM'
    # Add more custom logic if needed
    else:
        return 'Other'

# Apply function
dataset['Category'] = dataset.apply(categorize_flight, axis=1)
```

```python
eligibility_lookup = {
    'Short-haul AM': {'Tier 1': 5, 'Tier 2': 20, 'Tier 3': 60},
    'Mid-haul AM': {'Tier 1': 7, 'Tier 2': 25, 'Tier 3': 65},
    # Add more categories if needed
}
```

```python
dataset['Tier 1 %'] = dataset['Category'].map(lambda x: eligibility_lookup.get(x, {}).get('Tier 1', 0))
dataset['Tier 2 %'] = dataset['Category'].map(lambda x: eligibility_lookup.get(x, {}).get('Tier 2', 0))
dataset['Tier 3 %'] = dataset['Category'].map(lambda x: eligibility_lookup.get(x, {}).get('Tier 3', 0))
```

```python
result = dataset[['FLIGHT_NO', 'FLIGHT_TIME', 'ARRIVAL_REGION', 'Category', 'Tier 1 %', 'Tier 2 %', 'Tier 3 %']]
display(result.head())
```
